```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
score = pd.read_csv('data/score.csv', index_col=0)
base = pd.read_csv('data/base.csv', index_col=0)
count = pd.read_csv('data/count.csv', index_col=0)
inning = pd.read_csv('data/inning.csv', index_col=0)
telop_info = pd.concat([inning, base, count, score], axis=1)
telop_info
```

```python
row_before = telop_info.iloc[0]
get_score_scene = []
for row in telop_info.iterrows():
    if row[1]['score_1'] != row_before['score_1'] or row[1]['score_2'] != row_before['score_2']:
        get_score_scene.append(row[0])
    row_before = row[1]
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
filepath = "../game.mp4" # movie file
```

```python
def read_movie():
    movie = cv2.VideoCapture(filepath)
    if not movie.isOpened():
        print("can't open")
        exit()
    return movie
target_frame = 2000
movie = read_movie()
for i in range(target_frame):
    ret, frame = movie.read()
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(frame[40:165, 65:250], cv2.COLOR_BGR2RGB))
```

```python
print(movie.get(cv2.CAP_PROP_FRAME_COUNT))
print(movie.get(cv2.CAP_PROP_FPS))
```

```python
movie = read_movie()
score = cv2.imread("score.jpg")
i = 0
while True:
    ret, frame = movie.read()
    if i%60==0:
        score_diff = cv2.absdiff(frame[40:165, 65:250], score)
        if score_diff.mean() < 20:
            print(score_diff.mean())
            cv2.imwrite(f"scores/score{i}.jpg", frame[40:165, 65:250])
    i+=1
    if i == 10000:
        break
```

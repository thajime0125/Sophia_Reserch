```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
filepath = "../data/kochi_gameonly.mp4" # movie file
```

```python
def read_movie():
    movie = cv2.VideoCapture(filepath)
    if not movie.isOpened():
        print("can't open")
        exit()
    return movie
target_frame = 1000
movie = read_movie()
for i in range(target_frame):
    ret, frame = movie.read()
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
movie = read_movie()
score = cv2.imread("../data/score.jpg")
i = 0
while True:
    ret, frame = movie.read()
    if i%60==0:
        score_diff = cv2.absdiff(frame[5:95, 955:1275], score)
        if score_diff.mean() < 30:
            print(score_diff.mean())
            cv2.imwrite(f"../data/scores/score{i}.jpg", frame[5:95, 955:1275])
    i+=1
    if i == 259230:
        break
```
